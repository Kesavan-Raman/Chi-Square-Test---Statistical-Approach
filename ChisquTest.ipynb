```
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/srivatsan88/YouTubeLI/master/dataset/churn_data_st.csv")
```

```
data.head()
```

```
cont_table=pd.crosstab(data["gender"],data["Churn"])
print("contingency table:\n",cont_table)
```

```
obs_values=cont_table.values
print(obs_values)
```

```
import scipy.stats
p=scipy.stats.chi2_contingency(cont_table)
exp_values=p[3]
print('Expected values:\n',exp_values)
```

```
rows=len(cont_table.iloc[0:2,0])
columns=len(cont_table.iloc[0,0:2])
dof=(rows-1)*(columns-1)
print("degrees of freedom:\n",dof)
```

```
from scipy.stats import chi2
cal_val=sum([(o-e)**2./e for o,e in zip(obs_values,exp_values)])
cal_value=cal_val[0]+cal_val[1]
print("calculted value=\n",cal_value)
```

```
alpha=0.05
tab_value=chi2.ppf(q=1-alpha,df=dof)
print('Tabulated value=\n',tab_value)
```

```
if(cal_value>=tab_value):
  print('H0(null hypothesis) is rejected.i,e There is a significant relation between gender and the churning factor')
else:
  print('H0(null hypothesis) is accepted.i,e There is NO significant relation between gender and the churnring factor')
```
